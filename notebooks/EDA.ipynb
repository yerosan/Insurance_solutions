#### Summarized Insights:
 #### Total Premium Distribution by Province (Box Plot):

- Most provinces have similar premium distributions with a few outliers, especially in the Western Cape.
- The outliers indicate that certain policies in provinces like Western Cape have significantly higher premiums compared to the majority of policies.
 #### Regression between Total Premium and Total Claims:

- There is a weak positive correlation between TotalPremium and TotalClaims, meaning that as premiums increase, claims tend to rise, but the relationship is not strong.
- A cluster of low premiums shows wide variability in claims, with some high claims even for low premium amounts.
Outliers exist with extremely high claims for both low and high premiums, requiring further analysis to understand the risk dynamics.
