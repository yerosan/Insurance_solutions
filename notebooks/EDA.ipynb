### Data Comparison
Comparing Trends Over Geography:  the change in insurance cover type, premium, auto make, etc. 

```python
eda.correlation_matrix(cleand_data)
```

#### Correlation Matrix:
#### *Observation:**
- ***TotalPremium and TotalClaims:*** Weak positive correlation (0.12), indicating limited linear relationship.
- ***TotalPremium and CalculatedPremiumPerTerm:*** Strong correlation (0.64), suggesting a close relationship between these two features.
SumInsured has almost no correlation with TotalPremium or TotalClaims.
- **Key Insight:** The strongest linear relationship is between TotalPremium and CalculatedPremiumPerTerm, while the other variables show weak or no correlation with each other.
